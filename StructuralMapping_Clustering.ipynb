```python
import os
from Bio.PDB import PDBParser
import numpy as np

parser = PDBParser(QUIET=True)

def get_ca_coordinates(pdb_path, residue_list):
    structure = parser.get_structure("m", pdb_path)
    chain = next(structure.get_chains())
    ca_coords = []
    for res in chain:
        resnum = res.get_id()[1]
        if resnum in residue_list and "CA" in res:
            ca_coords.append((resnum, res["CA"].get_coord()))
    return ca_coords

# Directory containing AlphaFold PDBs
structure_dir = "/content/alphafold_structures"

coords_dict = {}
for uid, residues in sites_by_protein.items():
    pdb_file = f"{structure_dir}/{uid}.pdb"
    if not os.path.exists(pdb_file):
        print(f"Skipping {uid} — structure not found.")
        continue
    coords_dict[uid] = get_ca_coordinates(pdb_file, residues)
```

Cluster them in 3D to find spatial hotspots

```python
from sklearn.cluster import DBSCAN

def cluster_residues(ca_list, eps=12.0, min_samples=2):
    resnums, xyz = zip(*ca_list)
    xyz = np.vstack(xyz)
    labels = DBSCAN(eps=eps, min_samples=min_samples).fit_predict(xyz)
    clusters = {}
    for lbl, rn in zip(labels, resnums):
        if lbl == -1: continue   # noise
        clusters.setdefault(lbl, []).append(rn)
    return clusters

clusters_by_protein = {
    uid: cluster_residues(ca_list)
    for uid, ca_list in coords_dict.items()
}
```

Summarize and prioritize clusters

```python
for uid, clusters in clusters_by_protein.items():
    print(f"\n{uid} — found {len(clusters)} cluster(s):")
    for cid, resnums in clusters.items():
        print(f"  • Cluster {cid}: {resnums}")
```

```python

#Build a summary table
rows = []
for uid, clus in clusters_by_protein.items():
    #clusters of size ≥ 3
    big_clusters = [c for c in clus.values() if len(c) >= 3]
    rows.append({'uniprot_id': uid,'total_clusters': len(clus),'clusters_≥3': len(big_clusters),
        'largest_cluster_size': max((len(c) for c in clus.values()), default=0)})
df = pd.DataFrame(rows)

#Sort by number of big clusters, then by largest_cluster_size
df = df.sort_values(['clusters_≥3','largest_cluster_size'], ascending=False)
df.head(10)
```

```python
import py3Dmol

uid = "Q9Y5K6"
clusters = clusters_by_protein[uid]
pdb_path = f"/content/alphafold_structures/{uid}.pdb"

#Read PDB
with open(pdb_path) as f:
    pdb_str = f.read()

#Viewer setup
view = py3Dmol.view(width=700, height=500)
view.addModel(pdb_str, 'pdb')
view.setStyle({'cartoon': {'color': 'lightgrey', 'opacity': 0.8}})

#Define exactly three colors:
cluster_colors = {0: '#FF0000', 1: '#FFA500', 2: '#FFFF00'}

for cid, resnums in clusters.items():
    if len(resnums) < 3:
        continue
    color = cluster_colors[cid]
    print(f"Plotting cluster {cid} (n={len(resnums)}) in {color}")
    view.addStyle({'chain': 'A', 'resi': resnums},{'cartoon': {'color': color, 'opacity': 1.0},
         'stick':   {'radius': 0.2, 'color': color},'sphere':  {'scale': 0.2, 'color': color}})

view.addLabel(
  f"Cluster {cid}",
  { 'position': { 'resi': resnums[0], 'chain': 'A' },'backgroundColor': 'white','fontColor': color,
    'fontSize': 14 })

view.zoomTo()
view.setBackgroundColor('0xffffff')
view.render()
view.show()
```
